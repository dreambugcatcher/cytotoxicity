```python
from json import dump, load
from os import listdir
from pandas import DataFrame, read_excel, concat
from tqdm import tqdm
```

```python
with open('new_data_DTXSID.json', 'r') as file:
    new_data = load(file)
```

```python
data = DataFrame()

for file in tqdm(listdir('../toxval/dumps')):
    data = concat((
        data,
        read_excel(f'../toxval/dumps/{file}')
    ))
```

```python
new_data_smi = {}
for id, value in tqdm(new_data.items()):
    smi_data = data[data['DTXSID'] == id]['SMILES']
    if not smi_data.empty:
        smi = smi_data.iloc[0]
        new_data_smi[smi] = new_data_smi.get(smi, []) + value
    else:
        print(id)
```

```python
with open('new_data_smi_v3.json', 'w', encoding='utf-8') as file:
    dump(new_data_smi, file, indent=4)
```
